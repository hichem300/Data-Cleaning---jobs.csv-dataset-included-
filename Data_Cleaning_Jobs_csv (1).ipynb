##Vérification du formatage de date

```python
#On remarque que le format des colonnes contenant les dates c'est des string (object), on doit les convertir en datetime
#datetime dans Pandas Python permet de travailler plus aisément avec les dates et effectuer des opérations importantes

jobs_df.info()
```

```python
#Fonction pour verification du formatage de date

import dateutil

def is_valid_date_format(date_column):
      
  
    #Validates the date format in the given column.
    #Returns True if the date format is valid, False otherwise

  
    # Verify the date format using the dateutil.parser.parse function
    try:
        dateutil.parser.parse(date_column)
        return True
    except ValueError:
        return False

```

```python
column_dates=[40,41,42,44,45,85,87,11,81]
```

```python
for i in column_dates:
  (~jobs_df.iloc[:,i].apply(is_valid_date_format)).sum()
```

```python
column_date=[40,41,42,43,44,45,85,87]
column_date_heure=[11,81]
```

```python
#Conversion vers le bon formatage
for i in column_date: 
  jobs_df.iloc[:, i] = pd.to_datetime(jobs_df.iloc[:, i], format='%m/%d/%Y', errors='coerce').dt.strftime('%Y-%m-%d')
```

```python
for i in column_date_heure: 
  jobs_df.iloc[:, i] = pd.to_datetime(jobs_df.iloc[:, i], format='%m/%d/%Y %I:%M:%S %p', errors='coerce').dt.strftime('%Y-%m-%d %H:%M:%S')
```
